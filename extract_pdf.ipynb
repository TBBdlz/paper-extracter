```python
import PyPDF2
import os
import re
```

```python
pdf_data_dir = os.path.join(os.getcwd(), "pdf_data")
```

```python
def extract_th_abstract(pdf_text: str) -> str:
	th_abs_index = pdf_text.find('บทคัดย่อ')
	th_abs_len = len('บทคัดย่อ')
	th_keyword_index = pdf_text.find('คำสำคัญ')
	return pdf_text[th_abs_index + th_abs_len:th_keyword_index].strip()
```

```python
def extract_th_keyword(pdf_text: str) -> str:
    th_keyword_index = pdf_text.find('คำสำคัญ')
    eng_abs_index = pdf_text.find('Abstract')
    return pdf_text[th_keyword_index + len('คำสำคัญ'):eng_abs_index].strip()
```

```python
def extract_eng_abstract(pdf_text: str) -> str:
	eng_abs_index = pdf_text.find('Abstract')
	eng_abs_len = len('Abstract')
	return pdf_text[eng_abs_index + eng_abs_len:eng_keyword_index].strip()
```

```python
def extract_eng_keyword(pdf_text: str) -> str:
	eng_keyword_index = pdf_text.find('Keywords')
	eng_keyword_len = len('Keywords')
	th_intro_index = pdf_text.find('บทนำ')
	return pdf_text[eng_keyword_index + eng_keyword_len:th_intro_index].strip()
```

```python
#! code might look ugly :(
#? demo usage can be modified to fit your need
# loop throgh file in ./pdf_data
for file in os.listdir(pdf_data_dir):
	with open(os.path.join(pdf_data_dir, file), 'rb') as f:
		print(f"Start reading file: {file}")
		print("------------------------------------------")
		pdf_reader = PyPDF2.PdfReader(f)
		num_pages = len(pdf_reader.pages)
		for page in range(num_pages):
			pdf_page = pdf_reader.pages[page]
			pdf_text = pdf_page.extract_text()
			pdf_text = pdf_text.replace(' า', 'ำ')
			pdf_text = pdf_text.replace('�', 'ํ')
			if 'บทคัดย่อ' in pdf_text:
				print('----- บทคัดย่อ -----')
				th_abstract = extract_th_abstract(pdf_text)
				print(th_abstract)
			if 'คำสำคัญ' in pdf_text:
				print('----- คำสำคัญ ------')
				th_keyword = extract_th_keyword(pdf_text)
				print(th_keyword)
			if 'Abstract' in pdf_text:
				print('----- Abstract -----')
				eng_abs_index = pdf_text.find('Abstract')
				eng_keyword_index = pdf_text.find('Keywords')
				if eng_keyword_index == -1:
					eng_keyword_index = pdf_reader.pages[page + 1].extract_text().find('Keywords')
					eng_abstract = pdf_text[eng_abs_index:] + pdf_reader.pages[page + 1].extract_text()[:eng_keyword_index]
				else:
					eng_abstract = pdf_text[eng_abs_index]
				print(eng_abstract)
			if 'Keywords' in pdf_text:
				print('----- Keywords -----')
				eng_keyword = extract_eng_keyword(pdf_text)
				print(eng_keyword)
```
